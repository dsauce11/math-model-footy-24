```python
#now lets find the player
#how does player dataset looks like
df_players.head()
#lets look for laca and his ID to match with Events
laca_id = df_players.loc[df_players['lastName'] == 'Lacazette']['wyId'].values[0]
print(laca_id)
```

```python
#now we have his id. lets test it out and confirm
df_players[df_players['wyId'] == laca_id]
```

```python
#now lets get his shot seperated by foot
df_laca_events = df_shots_only[df_shots_only['playerId'] == laca_id]
df_laca_leet = df_laca_events

#left leg shots
lefty_shots = df_laca_events.loc[df_laca_events.apply (lambda x:{'id':401} in x.tags, axis = 1)]
#right leg shots
righty_shots = df_laca_events.loc[df_laca_events.apply (lambda x:{'id':402} in x.tags, axis = 1)]

l = [1] * len(lefty_shots)
l.extend([-1] * len(righty_shots))
```

```python
#test the hypothesis and show if Lacazeete is ambidextrous or not
from statsmodels.stats.descriptivestats import sign_test
test = sign_test(l, mu0 = 0)
pvalue = test[1]

if pvalue < 0.05:
    print("P-value amounts to", str(pvalue)[:5], "- We reject null hypothesis - Laca is not ambidextrous")
else:
    print("P-value amounts to", str(pvalue)[:5], " - We do not reject null hypothesis - Laca is ambidextrous")
```

```python
#Lets use T-tests
import pandas as pd
import numpy as np
import json
# plotting
import matplotlib.pyplot as plt
#opening data
import os
import pathlib
import warnings
pd.options.mode.chained_assignment = None
warnings.filterwarnings('ignore')
```

```python
#lets open/get the data
#events data
train = df_events_eng
#team data
with open('../data/wyscout/teams.json', 'r') as f:
    data = json.load(f)

df_wys_teams = pd.DataFrame(data)
df_wys_teams = df_wys_teams.rename(columns={"wyId":"teamId"})
df_wys_teams.info()
```

```python
#get corner data
corners = train.loc[train['subEventName']== 'Corner']
#count corner per team
corners_by_team = corners.groupby(['teamId']).size().reset_index(name='counts')
#merge with team name
summary = corners_by_team.merge(df_wys_teams[["name", "teamId"]], how = "left", on = ["teamId"])
#count corners by team by game
corners_by_game = corners.groupby(['teamId', "matchId"]).size().reset_index(name='counts')
#merge with team name
summary2 = corners_by_game.merge(df_wys_teams[["name", "teamId"]], how = "left", on = ["teamId"])
```

```python
#lets test out corners a team like arsenal gets
team_name= 'Arsenal'

Arsenal_corners = summary2.loc[summary2["name"] == 'Arsenal']["counts"]

def FormatFigure(ax):
    ax.legend(loc='upper left')
    ax.set_ylim(0,0.25)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylabel('')
    ax.set_xlabel('Corners')
    ax.set_ylabel('Proportion of games')
    ax.set_xticks(np.arange(0,21,step=1))


fig,ax1=plt.subplots(1,1)
ax1.hist(Arsenal_corners, np.arange(0.01,20.5,1), color='lightblue', edgecolor = 'white',linestyle='-',alpha=0.5, label=team_name, density=True,align='right')
FormatFigure(ax1)


mean = Arsenal_corners.mean()
std = Arsenal_corners.std()

print('Arsenal typically had %.2f  plus/minus %.2f corners per match in the 2017/18 season.'%(mean,std))
```

```python
from scipy.stats import ttest_1samp
t, pvalue = ttest_1samp(Arsenal_corners ,popmean=6)

print("The t-staistic is %.2f and the P-value is %.2f."%(t,pvalue))
if pvalue < 0.05:
    print("We reject null hypothesis - " + team_name + " typically take more than 6 corners per match.")
else:
    print("We cannot reject null hypothesis - " + team_name + " do not typically take more than 6 corners per match.")
```

```python
#two sided t-test
#lets compare liv and everton
liverpool_corners = summary2.loc[summary2["name"] == 'Liverpool']["counts"]
everton_corners = summary2.loc[summary2["name"] == 'Everton']["counts"]

mean = liverpool_corners.mean()
std = liverpool_corners.std()
print('Liverpool typically had %.2f plus/minus %.2f corners per match in the 2017/18 season.'%(mean,std))
std_error=std/np.sqrt(len(liverpool_corners))
print('The standard error in the number of corners per match is %.4f'%std_error)

mean = everton_corners.mean()
std = everton_corners.std()
print('Everton typically had %.2f plus/minus %.2f corners per match in the 2017/18 season.'%(mean,std))
std_error=std/np.sqrt(len(everton_corners))
print('The standard error in the number of corners per match is %.4f'%std_error)
```

```python
fig,ax=plt.subplots(1,1)
ax.hist(liverpool_corners, np.arange(0.01,15.5,1), color='red', edgecolor = 'white',linestyle='-',alpha=1.0, label="Liverpool", density=True,align='right')
ax.hist(everton_corners, np.arange(0.01,15.5,1), alpha=0.25, color='blue', edgecolor = 'black', label='Everton',  density=True,align='right')
FormatFigure(ax)
```

```python
# Here we test if Liverpool had a different average corners per game than Everton.
# We set the significance level at 0.05.


from scipy.stats import ttest_ind
t, pvalue  = ttest_ind(a=liverpool_corners, b=everton_corners, equal_var=True)

print("The t-staistic is %.2f and the P-value is %.2f."%(t,pvalue))
if pvalue < 0.05:
    print("We reject null hypothesis - Liverpool took different number of corners per game than Everton")
else:
    print("We cannot reject the null hypothesis that Liverpool took the same number of corners per game as Everton")
```
